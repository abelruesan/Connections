```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

Función que se encarga de generar la onda cuadrada dado el intervalo, el período y la amplitud de la onda.

```python
def ondaCuadrada(x, T, A):
    valores_onda_cuadrada = []

    if type(x) == list:
      for elemento in x:
        if 0 <= elemento % T < T / 2:
          valores_onda_cuadrada.append(A)
        elif T / 2 <= elemento % T <= T:
          valores_onda_cuadrada.append(-A)

      return valores_onda_cuadrada

    else:
      if 0 <= x % T < T / 2:
        return A

      elif T / 2 <= x % 2 <= T:
        return -A

```

```python
def aproximacionFourierOndaCuadrada(listaAbscisas, numTerminos, T, A):
    valores = []
    factor = 4 * A / math.pi

    for elemento in listaAbscisas:
        suma = 0
        for n in range(1, numTerminos * 2, 2):
            suma += (1 / n) * math.sin((2 * n * math.pi * elemento) / T)

        valor = factor * suma
        valores.append(valor)

    return valores
```

```python
def dibujaOndaCuadrada(intervalo, numTerminos, T, A):
  valores_x = list(np.linspace(intervalo[0], intervalo[1], 1000))
  valores_y = ondaCuadrada(valores_x, T, A)
  plt.plot(valores_x, valores_y, "r")
  valores_y_fourier = aproximacionFourierOndaCuadrada(valores_x, numTerminos, T, A)
  plt.plot(valores_x, valores_y_fourier, "b")
  plt.title(f"""Aproximación de una Serie de Fourier de una Onda Cuadrada (T={T} segundos, A={A}),
  en el intervalo {intervalo} usando {numTerminos} términos""")
  plt.xlabel("x (segundos)")
  plt.ylabel("y")
  plt.show()
```
